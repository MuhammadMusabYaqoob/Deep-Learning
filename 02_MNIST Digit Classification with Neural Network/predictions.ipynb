```python
import cv2
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings('ignore')
```

```python
# Load your trained model (if not already in memory)
model = tf.keras.models.load_model("models_saved/model.keras")
```

```python
# Load the image taken from your mobile
img = cv2.imread("images/one.png") 
```

```python
plt.imshow(img)
```

```python
# Convert to grayscale
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
gray.shape
```

```python
plt.imshow(gray)
```

```python
# Resize to 28x28
resized = cv2.resize(gray, (28, 28))
resized.shape
```

```python
# Normalize the pixel values
resized = resized / 255.0
```

```python
# Reshape to match model input shape
reshaped = resized.reshape(1, 28, 28, 1)
reshaped.shape
```

```python
# Predict
prediction = model.predict(reshaped)
print("Probabilities:", prediction)
print("Predicted Class:", prediction.argmax())
```
