## As image in RGB color space and has 3 channels (red, green, and blue).

### Let's resize the image

```python
from PIL import Image

img = Image.open('dog.jpg')
img_resized = img.resize((200, 200))
```

```python
# Save the resized image
img_resized.save('dog_image_resized.jpg')
```

```python
# displaying the resized image

img_res = mpimg.imread('dog_image_resized.jpg')
img_res_plot = plt.imshow(img_res)
plt.show()
```

### Image is resized in 200 x 200 pixels

```python
# Display the shape of the resized image
print(img_res.shape)
```

### We resized the image now we will convert into grayscale(1 channel)

```python
import cv2

grayscale_image = cv2.cvtColor(img_res, cv2.COLOR_RGB2GRAY)

# print the shape of the grayscale image
print(grayscale_image.shape)
```

```python
# display the grayscale image
plt.imshow(grayscale_image)
plt.show()
```

```python
# saving the grayscale image

cv2.imwrite('dog_grayscale_image.jpg', grayscale_image)
```
